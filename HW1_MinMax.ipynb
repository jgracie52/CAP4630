##Homework 1
Part A: 
Find three simple functions from R^2 to R and suitable points that are:

(1) a local minimum

(2) a local maximum

(3) neither a minimum nor maximum, but the gradient is zero at this point.

##Minimum Function

```
from mpl_toolkits import mplot3d
from matplotlib import cm
%matplotlib inline
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = plt.axes(projection='3d')


def f_1(x, y):
    return np.sqrt(x**2 + y ** 2)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z = f_1(X, Y)
ax1.contour3D(X, Y, Z, 50, cmap='binary');
ax1.scatter(0 , 0 , f_1(0,0),  color='green')
```

##Maximum Function

```
# Create the figure
fig2 = plt.figure()

# Add an axes
ax2 = fig2.add_subplot(111,projection='3d')

def f_2(x, y):
    return -(x**2 + y**2)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z = f_2(X, Y)

#Contour was making it difficult to see the point clearly
#ax2.contour3D(X, Y, Z, 50, cmap='binary'); 

# plot the surface
ax2.plot_surface(X, Y, Z, alpha=0.2)

# and plot the point 
ax2.scatter(0 , 0, f_2(0,0),  color='green')
```

##Saddle Point Function

```
# Create the figure
fig3 = plt.figure()

# Add an axes
ax3 = fig3.add_subplot(111,projection='3d')

def f_3(x, y):
    return x**3 + y**3

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z = f_3(X, Y)
ax3.contour3D(X, Y, Z, 50, cmap='binary');
ax3.scatter(0 , 0 , f_3(0,0),  color='red')
```

##Quiver Gradients

###Gradient of Min Function

```
#Number of values produced by linspace was reduced to 20 to allow for easier viewing of the quiver plot
x,y = meshgrid(np.linspace(-10, 10, 20),np.linspace(-10, 10, 20))
f = np.sqrt(x**2 + y ** 2)
dy,dx = np.gradient(f, .2, .2)

fig, ax = plt.subplots()
ax.quiver(x, y, dx, dy)
plt.show()
```

###Gradient of Max Function

```
#Number of values produced by linspace was reduced to 20 to allow for easier viewing of the quiver plot
x,y = meshgrid(np.linspace(-10, 10, 20),np.linspace(-10, 10, 20))
f = -(x**2 + y ** 2)
dy,dx = np.gradient(f, .2, .2)

fig, ax = plt.subplots()
ax.quiver(x, y, dx, dy)
plt.show()
```

```
#Number of values produced by linspace was reduced to 20 to allow for easier viewing of the quiver plot
x,y = meshgrid(np.linspace(-10, 10, 20),np.linspace(-10, 10, 20))
f = x**3 + y ** 3
dy,dx = np.gradient(f, .2, .2)

fig, ax = plt.subplots()
ax.quiver(x, y, dx, dy)
plt.show()
```
